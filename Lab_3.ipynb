FOR SIMPLE LINEAR REGRESSION

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample dataset (x = input feature, y = target variable)
x = np.array([1, 2, 3, 4, 5])  # Feature
y = np.array([2, 4, 6, 8, 10])  # Target (y = 2x)

# Initialize parameters
theta_0 = 0  # Intercept
theta_1 = 0  # Slope
alpha = 0.01  # Learning rate
epochs = 1000  # Number of iterations
m = len(x)  # Number of data points

# Gradient Descent Algorithm
cost_history = []

for _ in range(epochs):
    # Predictions
    h = theta_0 + theta_1 * x

    # Compute gradients
    d_theta0 = (1/m) * np.sum(h - y)
    d_theta1 = (1/m) * np.sum((h - y) * x)

    # Update parameters
    theta_0 -= alpha * d_theta0
    theta_1 -= alpha * d_theta1

    # Compute cost
    cost = (1/(2*m)) * np.sum((h - y) ** 2)
    cost_history.append(cost)

# Final parameters after training
print(f"Final Parameters: θ0 = {theta_0}, θ1 = {theta_1}")

# Plot cost function decrease over epochs
plt.plot(cost_history)
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.title("Cost Function Decrease")
plt.show()
```

FOR MULTI LINEAR REGRESSION

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample dataset (X = input features, y = target variable)
X = np.array([
    [1, 2100, 3],   # House 1: 2100 sqft, 3 bedrooms
    [1, 1600, 2],   # House 2: 1600 sqft, 2 bedrooms
    [1, 2400, 3],   # House 3: 2400 sqft, 3 bedrooms
    [1, 1410, 2],   # House 4: 1410 sqft, 2 bedrooms
    [1, 3000, 4]    # House 5: 3000 sqft, 4 bedrooms
])  # Added 1s in the first column for bias term (θ0)
y = np.array([399900, 329900, 369000, 232000, 539900])  # House prices

# Initialize parameters (theta), learning rate, and iterations
theta = np.zeros(X.shape[1])  # Initialize θ0, θ1, θ2 to 0
alpha = 0.00000001  # Learning rate (tuned to a small value)
epochs = 1000  # Number of iterations
m = len(y)  # Number of training examples

# Gradient Descent Algorithm
cost_history = []

for _ in range(epochs):
    # Predictions (h = X * theta)
    h = X.dot(theta)

    # Compute gradients
    gradients = (1/m) * X.T.dot(h - y)

    # Update theta values
    theta -= alpha * gradients

    # Compute cost function
    cost = (1/(2*m)) * np.sum((h - y) ** 2)
    cost_history.append(cost)

# Final parameters after training
print(f"Final Parameters: {theta}")

# Plot cost function decrease over epochs
plt.plot(cost_history)
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.title("Cost Function Decrease in Multiple Linear Regression")
plt.show()
```
